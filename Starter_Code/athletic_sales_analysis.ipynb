```python
# Import Libraries and Dependencies
import pandas as pd
from pathlib import Path 
```

### 1. Combine and Clean the Data
#### Import CSVs

```python
new_df = Path('Resources/athletic_sales_2020.csv')
sales_df = Path('Resources/athletic_sales_2021.csv')
```

```python
# Read the CSV files into DataFrames.
new_df = pd.read_csv('Resources/athletic_sales_2020.csv')
new_df.head()
```

```python
sales_df = pd.read_csv('Resources/athletic_sales_2021.csv')
sales_df.head()
```

```python
# Display the 2021 sales DataFrame
sales_df = pd.DataFrame(sales_df)
sales_df
```

#### Combine the sales data by rows.

```python
# Combine the 2020 and 2021 sales DataFrames on the rows and reset the index.
joined_data = pd.concat([new_df, sales_df], axis="rows", join="inner")
joined_data
```

```python
joined_data['invoice_date']= pd.to_datetime(joined_data['invoice_date'], errors='coerce')

joined_data= joined_data.dropna(subset=['invoice_date']).reset_index(drop=True)
joined_data.head(20)
```

### 2. Determine which Region Sold the Most Products

#### Using `groupby`

```python
joined_data
```

```python
joined_data["Sum"] = joined_data["units_sold"] + joined_data["total_sales"]
joined_data
```

```python
def custom_sum(x):
    return x.sum()
```

```python
updated_df = joined_data.groupby(["region", "state", "city"]).apply(lambda x: pd.Series({"sum": custom_sum(x["Total_Products_Sold"])}))
updated_df.head()
```

```python
# Show the number products sold for region, state, and city.
# Rename the sum to "Total_Products_Sold".

# Show the top 5 results.

updated_df = joined_data.groupby(["region", "state", "city"])[["Sum"]].size()
updated_df.head(10)
```

#### Using `pivot_table`

```python
# Show the number products sold for region, state, and city.
pi_table = pd.pivot_table(joined_data, 
                                   index=['region','state', 'city'],
                                   values='Sum',
                                   aggfunc='sum')
pi_table.head()

# Rename the "units_sold" column to "Total_Products_Sold"


# Show the top 5 results.

updated_pi_table = pi_table.rename(columns={"Sum": "Total_Products_Sold"})

updated_pi_table.sort_values(by=["Total_Products_Sold"], ascending=False).head(5)
```

### 6. Determine the Day with the Most Women's Athletic Footwear Sales

```python
women_df
```

```python
# Create a pivot table with the 'invoice_date' column is the index, and the "total_sales" as the values.


# Optional: Rename the "total_sales" column to "Total Sales"


# Show the table.

new_women_df = pd.pivot_table(women_df, 
                                 index=["invoice_date"],
                                 values='total_sales')
new_women_df.head(10)
```

```python
bins = [0, 1, 2, 3, 4, 5, 6]


updated_new_women_df= new_women_df.resample('D').count()


updated_new_women_df.sort_values(by="total_sales", ascending=False).head(10)
```

```python
# Resample the pivot table into daily bins, and get the total sales for each day.


# Sort the resampled pivot table in descending order on "Total Sales".

bins = [0, 1, 2, 3, 4, 5, 6]


updated_new_women_df= new_women_df.resample('D').count()


updated_new_women_df.sort_values(by="total_sales", ascending=False).head(10)
```

```python
# Resample the pivot table into weekly bins, and get the total sales for each week.

# Sort the resampled pivot table in descending order on "Total Sales".

updated_women_df= new_women_df.resample('W').count()


updated_women_df.sort_values(by="total_sales", ascending=False).head(10)
```
